```python
import os
import shutil
from tqdm import tqdm

root = r"C:\Users\ipl\.cache\kagglehub\datasets\awsaf49\asvpoof-2019-dataset\versions\1\LA\LA"

audio_dirs = {
    "dev": os.path.join(root, "ASVspoof2019_LA_dev", "flac"),
    "eval": os.path.join(root, "ASVspoof2019_LA_eval", "flac"),
}

protocol_dir = os.path.join(root, "ASVspoof2019_LA_cm_protocols")

protocol_files = [
    "ASVspoof2019.LA.cm.dev.trl.txt",
    "ASVspoof2019.LA.cm.eval.trl.txt"
]

real_out = os.path.join(root, "REAL")
fake_out = os.path.join(root, "FAKE")

os.makedirs(real_out, exist_ok=True)
os.makedirs(fake_out, exist_ok=True)

# ✔️ 1. label map (dev + eval만)
label_map = {}

for pf in protocol_files:
    path = os.path.join(protocol_dir, pf)

    with open(path, "r") as f:
        for line in f:
            parts = line.strip().split()

            file_id = parts[1]
            label = parts[-1]

            label_map[file_id] = 0 if label == "bonafide" else 1

# ✔️ 2. train 이후 번호 이어받기 (핵심)
real_files = [f for f in os.listdir(real_out) if f.endswith(".flac")]
fake_files = [f for f in os.listdir(fake_out) if f.endswith(".flac")]

real_count = len(real_files) + 1
fake_count = len(fake_files) + 1

print(f"이어받은 REAL 시작 번호: {real_count}")
print(f"이어받은 FAKE 시작 번호: {fake_count}")

# ✔️ 3. dev + eval만 처리
for split_name, audio_dir in audio_dirs.items():

    files = [f for f in os.listdir(audio_dir) if f.endswith(".flac")]

    print(f"\n[{split_name}] 처리 시작: {len(files)}개")

    for file in tqdm(files):
        file_id = file.replace(".flac", "")
        src = os.path.join(audio_dir, file)

        if file_id not in label_map:
            continue

        label = label_map[file_id]

        if label == 0:
            dst = os.path.join(real_out, f"{split_name}_real{real_count}.flac")
            real_count += 1
        else:
            dst = os.path.join(fake_out, f"{split_name}_fake{fake_count}.flac")
            fake_count += 1

        shutil.move(src, dst)

print("\n완료!")
print(f"REAL 총: {real_count-1}")
print(f"FAKE 총: {fake_count-1}")
```

```python
import os
import subprocess
from tqdm import tqdm

SRC_ROOT = r"C:\Users\ipl\Downloads\train-clean-100\LibriSpeech\train-clean-100"
DST_ROOT = r"C:\Users\ipl\Downloads\train-clean-100\LibriSpeech\train-clean-100\REAL3"

os.makedirs(DST_ROOT, exist_ok=True)

# ===== flac 수집 =====
flac_files = []
for root, _, files in os.walk(SRC_ROOT):
    for f in files:
        if f.lower().endswith(".flac"):
            flac_files.append(os.path.join(root, f))

print("총 FLAC:", len(flac_files))

# ===== 라벨 시작 =====
idx = 1
success = 0
fail = 0

for src in tqdm(flac_files, desc="Converting"):
    try:
        out_name = f"real3_{idx:08d}.wav"
        dst = os.path.join(DST_ROOT, out_name)

        cmd = [
            "ffmpeg",
            "-y",
            "-i", src,
            "-ar", "16000",
            "-ac", "1",
            dst
        ]

        result = subprocess.run(cmd, capture_output=True, text=True)

        if result.returncode == 0 and os.path.exists(dst):
            success += 1
            idx += 1
        else:
            fail += 1

    except Exception:
        fail += 1

print("\n===== 완료 =====")
print("성공:", success)
print("실패:", fail)
print("저장 위치:", DST_ROOT)
```

```python
import os

ROOT = r"C:\Users\ipl\Desktop\test\DATASET"

def print_tree(path, indent=""):
    try:
        items = os.listdir(path)
    except Exception:
        return

    folders = []
    files = []

    for item in items:
        full = os.path.join(path, item)
        if os.path.isdir(full):
            folders.append(item)
        else:
            files.append(item)

    # 파일 개수 출력 (이름은 안 출력)
    print(f"{indent}[FILES: {len(files)}]")

    # 폴더 탐색
    for folder in sorted(folders):
        full_path = os.path.join(path, folder)
        print(f"{indent}{folder}/")
        print_tree(full_path, indent + "    ")

print("FINAL_DATASET STRUCTURE\n")
print_tree(ROOT)
```

```python
import os
import random
import shutil
from math import floor

BASE = r"C:\Users\ipl\Desktop\test\FINAL_DATASET\FINAL"

REAL_SRC = os.path.join(BASE, "REAL")
FAKE_SRC = os.path.join(BASE, "FAKE")

OUT_BASE = BASE.replace("FINAL", "")

SPLITS = {
    "TRAIN": 0.8,
    "VAL": 0.1,
    "EVAL": 0.1
}

random.seed(42)

# --------------------------
# 1. 파일 리스트
# --------------------------
def get_files(path):
    return [
        os.path.join(path, f)
        for f in os.listdir(path)
        if f.endswith(".flac") or f.endswith(".wav")
    ]

real_files = get_files(REAL_SRC)
fake_files = get_files(FAKE_SRC)

random.shuffle(real_files)
random.shuffle(fake_files)

# --------------------------
# 2. split
# --------------------------
def split(files):
    n = len(files)
    n_train = floor(n * SPLITS["TRAIN"])
    n_val = floor(n * SPLITS["VAL"])

    train = files[:n_train]
    val = files[n_train:n_train+n_val]
    test = files[n_train+n_val:]

    return train, val, test


r_train, r_val, r_test = split(real_files)
f_train, f_val, f_test = split(fake_files)

splits = {
    "TRAIN": (r_train, f_train),
    "VAL": (r_val, f_val),
    "EVAL": (r_test, f_test)
}

# --------------------------
# 3. copy
# --------------------------
def copy(files, dst, prefix):
    os.makedirs(dst, exist_ok=True)

    for i, f in enumerate(files):
        ext = os.path.splitext(f)[1]
        out = os.path.join(dst, f"{prefix}_{i:06d}{ext}")
        shutil.copy2(f, out)

# --------------------------
# 4. 실행
# --------------------------
for name, (r, f) in splits.items():
    print(f"\n[{name}]")
    
    copy(r, os.path.join(OUT_BASE, name, "REAL"), "R")
    copy(f, os.path.join(OUT_BASE, name, "FAKE"), "F")

    print("REAL:", len(r))
    print("FAKE:", len(f))

print("\nDONE")
```

```python
import os
import random
import shutil
from pathlib import Path

SRC_ROOT = r"C:\Users\ipl\Desktop\test\FINAL_DATASET\FINAL"
DST_ROOT = r"C:\Users\ipl\Desktop\test\FINAL_DATASET"

CLASSES = ["REAL", "FAKE"]

# 비율
TRAIN_RATIO = 0.8
VAL_RATIO = 0.1
TEST_RATIO = 0.1

random.seed(42)

def collect_files(folder):
    files = []
    for root, _, fs in os.walk(folder):
        for f in fs:
            if f.endswith(".flac") or f.endswith(".wav"):
                files.append(os.path.join(root, f))
    return files

def safe_move(files, dst_dir):
    os.makedirs(dst_dir, exist_ok=True)
    for f in files:
        try:
            shutil.copy2(f, os.path.join(dst_dir, os.path.basename(f)))
        except Exception as e:
            print("ERROR:", f, e)

for cls in CLASSES:
    src = os.path.join(SRC_ROOT, cls)
    files = collect_files(src)
    random.shuffle(files)

    n = len(files)
    train_end = int(n * TRAIN_RATIO)
    val_end = train_end + int(n * VAL_RATIO)

    train_files = files[:train_end]
    val_files = files[train_end:val_end]
    test_files = files[val_end:]

    print(f"\n[{cls}]")
    print("total:", n)
    print("train:", len(train_files))
    print("val:", len(val_files))
    print("test:", len(test_files))

    safe_move(train_files, os.path.join(DST_ROOT, "TRAIN", cls))
    safe_move(val_files, os.path.join(DST_ROOT, "VAL", cls))
    safe_move(test_files, os.path.join(DST_ROOT, "TEST", cls))

print("\nDONE SPLIT")
```

```python
import os
from pydub import AudioSegment
from tqdm import tqdm

SRC_ROOT = r"C:\Users\ipl\Desktop\test\DATASET"
DST_ROOT = SRC_ROOT  # 같은 폴더에 wav 추가 생성

def convert_flac_to_wav(src_path, dst_path):
    audio = AudioSegment.from_file(src_path, format="flac")
    audio = audio.set_frame_rate(16000).set_channels(1)
    audio.export(dst_path, format="wav")

# 1. FLAC 수집
flac_files = []
for root, _, files in os.walk(SRC_ROOT):
    for f in files:
        if f.lower().endswith(".flac"):
            flac_files.append(os.path.join(root, f))

print("FLAC 개수:", len(flac_files))

# 2. 변환 (원본 유지 + wav 생성)
for f in tqdm(flac_files):
    try:
        wav_path = f.replace(".flac", ".wav")

        # 이미 wav 있으면 스킵
        if os.path.exists(wav_path):
            continue

        convert_flac_to_wav(f, wav_path)

    except Exception as e:
        print("ERROR:", f, e)

print("DONE: FLAC → WAV (KEEP ORIGINAL)")
```

```python
import os
import numpy as np
from pathlib import Path
from collections import defaultdict

TARGET_SR = 16000
MIN_DURATION = 0.5
MAX_DURATION = 30.0

def check_file(fpath):
    try:
        import librosa
        y, sr = librosa.load(fpath, sr=None, duration=MAX_DURATION + 1)
        duration = len(y) / sr
        
        issues = []
        if sr != TARGET_SR:
            issues.append(f"sr_mismatch={sr}")
        if duration < MIN_DURATION:
            issues.append(f"too_short={duration:.2f}s")
        if duration > MAX_DURATION:
            issues.append(f"too_long={duration:.2f}s")
        
        return {
            'path': fpath,
            'ok': len(issues) == 0,
            'sr': sr,
            'duration': duration,
            'issues': issues
        }
    except Exception as e:
        return {
            'path': fpath,
            'ok': False,
            'sr': None,
            'duration': None,
            'issues': [f"read_error"]
        }


def print_dataset_verification(DATASET_ROOT):
    print("표 1. 데이터셋 무결성 검증 결과")
    print()
    
    splits = ['train', 'val', 'test']
    classes = ['real', 'fake']
    results_data = []
    
    grand_total = 0
    grand_ok = 0
    grand_issues = 0
    all_issue_log = []
    
    for split in splits:
        for cls in classes:
            split_cls_path = os.path.join(DATASET_ROOT, split, cls)
            
            if not os.path.exists(split_cls_path):
                continue
            
            files = sorted(Path(split_cls_path).glob('*.wav'))
            total = len(files)
            
            if total == 0:
                continue
            
            ok_count = 0
            issue_count = 0
            durations = []
            sr_counter = defaultdict(int)
            issue_types = defaultdict(int)
            issue_log = []
            
            for fpath in files:
                r = check_file(str(fpath))
                
                if r['duration'] is not None:
                    durations.append(r['duration'])
                
                if r['sr'] is not None:
                    sr_counter[r['sr']] += 1
                
                if r['ok']:
                    ok_count += 1
                else:
                    issue_count += 1
                    for iss in r['issues']:
                        key = iss.split('=')[0]
                        issue_types[key] += 1
                    
                    if len(issue_log) < 20:
                        issue_log.append({
                            'file': os.path.basename(r['path']),
                            'issues': r['issues']
                        })
            
            grand_total += total
            grand_ok += ok_count
            grand_issues += issue_count
            all_issue_log.extend(issue_log)
            
            d = np.array(durations) if durations else np.array([0])
            dur_str = f"{d.min():.2f}s - {d.mean():.2f}s - {d.max():.2f}s"
            tot_hours = d.sum() / 3600
            
            sr_info = ", ".join(
                f"{sr//1000}kHz"
                for sr in sorted(sr_counter.keys())
            ) if sr_counter else "N/A"
            
            issue_str = "없음"
            if issue_types:
                issue_str = ", ".join(
                    f"{k} {v}개"
                    for k, v in sorted(
                        issue_types.items(),
                        key=lambda x: -x[1]
                    )
                )
            
            results_data.append({
                'split_cls': f"{split}/{cls}",
                'total': total,
                'ok': ok_count,
                'issues': issue_count,
                'dur_str': dur_str,
                'tot_hours': tot_hours,
                'sr': sr_info,
                'issue_info': issue_str
            })
    
    if grand_total == 0:
        print("검증할 파일이 없습니다.")
        print(f"경로 확인: {DATASET_ROOT}")
        return
    
    # 테이블 출력
    for r in results_data:
        print(f"{r['split_cls']:12s} | "
              f"파일 {r['total']:,}개 (정상 {r['ok']:,}, 문제 {r['issues']:,}) | "
              f"길이 {r['dur_str']} | "
              f"총 {r['tot_hours']:.2f}시간 | "
              f"SR {r['sr']} | "
              f"문제: {r['issue_info']}")
    
    print()
    print("표 2. 데이터셋 검증 최종 요약")
    print()
    
    print(f"전체 파일     : {grand_total:,}개")
    if grand_total > 0:
        print(f"정상 파일     : {grand_ok:,}개 ({grand_ok/grand_total*100:.2f}%)")
        print(f"문제 파일     : {grand_issues:,}개 ({grand_issues/grand_total*100:.3f}%)")
    print()
    
    if all_issue_log:
        print("문제 파일 목록:")
        for item in all_issue_log[:20]:
            issues_str = ", ".join(item['issues'])
            print(f"  - {item['file']}: {issues_str}")
        print()
    
    if grand_total == 0:
        status = "검증할 파일이 없습니다."
    elif grand_issues == 0:
        status = "모든 오디오 파일이 품질 기준을 만족한다."
    elif grand_issues / grand_total < 0.02:
        status = ("소수의 이상 파일이 존재하나 전체 대비 비율이 2% 미만이며, "
                  "학습 과정에서 자동 제외가 가능하므로 학습에 문제 없다.")
    else:
        status = ("이상 파일 비율이 2% 이상으로 상대적으로 높게 확인되었으므로 "
                  "재전처리를 권장한다.")
     
    print(f"검증 결과: {status}")


if __name__ == "__main__":
    DATASET_ROOT = r"C:\Users\ipl\Desktop\test\DATASET"
    print_dataset_verification(DATASET_ROOT)
```
